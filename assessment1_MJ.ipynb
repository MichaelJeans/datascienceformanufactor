```python
##Couple of filters to remove some table noise
filtered_corr = corr[corr > 0.5]
filtered_corr = filtered_corr - matrix
filtered_corr[np.isnan(filtered_corr)] = 0
filtered_corr
# strong correlation
strong_corr = filtered_corr[filtered_corr >= 0.70]
strong_corr = strong_corr.dropna(how="all")
strong_corr = strong_corr.dropna(how="all", axis = 1)
strong_corr
```

```python
# Moderate Correlation
moderate_corr = filtered_corr[filtered_corr < 0.70]
moderate_corr = moderate_corr.replace(0, np.nan)
moderate_corr = moderate_corr.dropna(how="all")
moderate_corr = moderate_corr.dropna(how="all", axis = 1)
moderate_corr
```
